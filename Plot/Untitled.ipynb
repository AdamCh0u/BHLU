```python
fig, axes = plt.subplots(9,4, figsize=(30, 45))
fig.subplots_adjust(hspace=0.3, wspace=0.1)
i=0
for ax in axes.flat:
    df = pd.read_excel("各区域1992-2015各年用地面积.xls",sheet_name = "ID_{}".format(i))
    df = df.set_index('Unnamed: 0')
    total = sum(df.iloc[:,0])
    if 1 in list(df.index):
        bare = df.loc[1,:]/total
    else:
        bare = [0*i for i in range(24)]
    if 2 in list(df.index):
        fore = df.loc[2,:]/total
    else:
        fore = [0*i for i in range(24)]
    if 6 in list(df.index):
        argr = df.loc[6,:]/total
    else:
        argr = [0*i for i in range(24)]
    if 7 in list(df.index):
        urbn = df.loc[7,:]/total
    else:
        urbn = [0*i for i in range(24)]
        
    x = list(df.columns)
    y = np.vstack([bare, fore, argr, urbn])
    
    pal = ["blue","g","gold","orangered"]

    labels = ["Bare", "Forest", "Agriculture","Urban"]
    ax.stackplot(x, y,labels =labels,colors=pal,alpha = 0.8)
    ax.set_ylim(0,1)
    ax.set_xlim(1992,2015)
    ax.legend(loc ='upper left')
    ax.set_title("{}".format(i))
    i+=1
```
