```python
class Vividict(dict):
	def __missing__(self, key):
		value = self[key] = type(self)()
		return value
```

```python
dic = Vividict()
for i in range(df.shape[0]):
    FID = int(df.iloc[i,5])
    year = int(df.iloc[i,6])
    cls = int(df.iloc[i,1].split("_")[1])
    dic[FID][year][cls]=dict(df.iloc[i,2:5])
```
